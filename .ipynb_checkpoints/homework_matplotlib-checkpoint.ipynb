## Task 1

Graph the Gaussian function with at least three different values of sigma. The plots must have:

- Both axes identified
- Title
- Legend

```python
# Defining values for sigma
sigma1 = 1
sigma2 = 0.5
sigma3 = 0.25
```

```python
# Creating x-values to plot
x_values = np.linspace(-5, 5, 1000)
```

```python
# Evaluating the Gaussian function using sigma
tempCalc1 = np.exp(-0.5*(x_values/sigma1)**2)
tempCalc2 = sigma1*(2*np.pi)**(1/2)
y_values1 = tempCalc1/tempCalc2

tempCalc1 = np.exp(-0.5*(x_values/sigma2)**2)
tempCalc2 = sigma2*(2*np.pi)**(1/2)
y_values2 = tempCalc1/tempCalc2

tempCalc1 = np.exp(-0.5*(x_values/sigma3)**2)
tempCalc2 = sigma3*(2*np.pi)**(1/2)
y_values3 = tempCalc1/tempCalc2
```

```python
# Plotting the values
plt.plot(x_values, y_values1, linestyle="-", label="$\sigma=1.00$", color="black")
plt.plot(x_values, y_values2, linestyle="-", label="$\sigma=0.50$", color="red")
plt.plot(x_values, y_values3, linestyle="-", label="$\sigma=0.25$", color="orange")

# Adding the legend
plt.legend()

# Adding the names
plt.title("Gaussian function for different $\sigma$ values")
plt.xlabel("x-values")
plt.ylabel("g(x)")

# Grid and show
plt.grid()
plt.show()
```

## Task 2

The archive data/temperature-central-park.dat contains mean temperature for Central Park, New York during January 2021. The values for temperature are recorded in Fahrenheit.

- Read the dataset using numpy.loadtxt.
- Convert the temperature values from Fahrenheit to Celsius. Since we will not be using the Fahrenheit values, you can overwrite the array with the new values.
- Visualize the temperature data with the graph type that you think is the most appropriate.
- Make sure to follow the best practices for generating plots (annotated axes, units, legend, title, etc).
- Save your figure.

```python
# Loading the data in Fahrenheit
temperature_data_file = Path(".")/"temperature-central-park.dat"
temperature_data      = np.loadtxt(temperature_data_file)

print(temperature_data[:5,:])
```

```python
# Converting the values to Celsius degrees
temperature_data[:,1] = (5/9)*(temperature_data[:,1] - 32)

print(temperature_data[:5,:])
```

```python
# Visualizing the daily temperature and setting 'filename'
filename = Path(".")/"temperature_celsius.png"

plt.plot(temperature_data[:,0], temperature_data[:,1], linestyle="-", color="orange")

# Adding the names
plt.title("Daily temperature in Central Park, New York. January 2021.", fontsize=10)
plt.xlabel("Day")
plt.ylabel("Temperature (Â°C)")

# Setting the axis
plt.axis([0, 32, -8, 15])
plt.xticks([1,5,10,15,20,25,31], [1,5,10,15,20,25,31])

# Grid and show
plt.grid()
plt.savefig(filename, dpi=300)
plt.show()
```
